```python
import manim as mn
from manim import *

config.media_width = "75%"
config.verbosity = "WARNING"

print(mn.__version__)
```

```python
%%manim -qm DiagramProduct

class DiagramProduct(Scene):
    def construct(self):
        self.camera.background_color = WHITE

        H1_prod = [MathTex(r"\mathcal{H}_{01}").set_color(BLACK).scale(2).move_to(LEFT + UP*1.5), Dot(UP*1.5).set_color(BLACK).scale(0.5), MathTex(r"\mathcal{H}_{01}").set_color(BLACK).scale(2).move_to(RIGHT + UP*1.5)]
        self.play(*[Create(mob) for mob in H1_prod], run_time = 0.5)
        self.wait()

        #Define diagrams
        line1 = Line([4, 0, 0], [2, 0, 0], stroke_width = 5).set_color(BLACK)
        line2 = Line([2, 0, 0], [0, 1, 0], stroke_width = 5).set_color(BLACK)
        line3 = Line([2, 0, 0], [0, -1, 0], stroke_width = 5).set_color(BLACK)
        diagram1_1 = VGroup(line1, line2, line3).move_to(LEFT*3)
        diagram1_2 = diagram1_1.copy()

        line1 = Line([0, 0, 0], [2, 0, 0], stroke_width = 5).set_color(BLACK)
        line2 = Line([2, 0, 0], [4, 1, 0], stroke_width = 5).set_color(BLACK)
        line3 = Line([2, 0, 0], [4, -1, 0], stroke_width = 5).set_color(BLACK)
        diagram2_1 = VGroup(line1, line2, line3).move_to(RIGHT*3)
        diagram2_2 = diagram2_1.copy()

        plus_sign_1 = MathTex("+").set_color(BLACK).move_to(ORIGIN).scale(2)
        parenthesis_l_1 = MathTex("(").set_color(BLACK).move_to(LEFT*6).scale(5)
        parenthesis_r_1 = MathTex(")").set_color(BLACK).move_to(RIGHT*6).scale(5)

        plus_sign_2 = MathTex("+").set_color(BLACK).move_to(ORIGIN).scale(2)
        parenthesis_l_2 = MathTex("(").set_color(BLACK).move_to(LEFT*6).scale(5)
        parenthesis_r_2 = MathTex(")").set_color(BLACK).move_to(RIGHT*6).scale(5)

        H1_set1 = VGroup(diagram1_1, diagram2_1, plus_sign_1, parenthesis_l_1, parenthesis_r_1).scale(0.4).move_to(LEFT*3 + UP*3.5)
        H1_set2 = VGroup(diagram1_2, diagram2_2, plus_sign_2, parenthesis_l_2, parenthesis_r_2).scale(0.4).move_to(RIGHT*3 + UP*3.5)
        self.play(*[mob.animate.shift(UP*2) for mob in H1_prod], Transform(H1_prod[0], H1_set1), Transform(H1_prod[2], H1_set2))

        diagram1_1copy = diagram1_1.copy()
        diagram2_1copy = diagram2_1.copy()
        diagram1_2copy = diagram1_2.copy()
        diagram2_2copy = diagram2_2.copy()

        self.wait()

        plus_sign_1 = MathTex("+", stroke_width = 3).set_color(BLACK).scale(1).move_to(LEFT+UP)
        plus_sign_2 = MathTex("+", stroke_width = 3).set_color(BLACK).scale(1).move_to(RIGHT*5 + UP)
        plus_sign_3 = MathTex("+", stroke_width = 3).set_color(BLACK).scale(1).move_to(DOWN*1 + LEFT*6)
        plus_sign_4 = MathTex("+", stroke_width = 3).set_color(BLACK).scale(1).move_to(DOWN*1 )

        self.play(diagram1_1.animate.move_to(LEFT*5 +UP*1), diagram1_2.animate.move_to(LEFT*3 +UP*1), Create(plus_sign_1), run_time = 1)
        self.play(diagram1_1copy.animate.move_to(RIGHT +UP*1), diagram2_2.animate.move_to(RIGHT*3 +UP*1), Create(plus_sign_2), run_time = 1)
        self.play(diagram2_1.animate.move_to(LEFT*4 +DOWN*1), diagram1_2copy.animate.move_to(LEFT*2 +DOWN*1), Create(plus_sign_3), run_time = 1)
        self.play(diagram2_1copy.animate.move_to(RIGHT*2 +DOWN*1), diagram2_2copy.animate.move_to(RIGHT*4 +DOWN*1), Create(plus_sign_4), run_time = 1)

```
