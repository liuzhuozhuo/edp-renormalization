```python
import matplotlib.pyplot as plt
from matplotlib.path import Path
import matplotlib.patches as patches

# Control points: start, control1, control2, end
verts = [
    (0., 0.),     # P0: Start
    (1.25, 1.),     # P1: Control point 1
    (2.5, 0.),    # P2: Control point 2
    (1.75, -1.),     # P3: End
    (1., 0.),     # P4: Control point 2
    (1.5, 1.),     # P5: Control point 1
    (2.75, 0.),     # P6: End     # P8: End
]

# Define path codes
codes = [
    Path.MOVETO,   # Move to P0
    Path.CURVE4,   # Bézier control point 1
    Path.CURVE4,   # Bézier control point 2
    Path.CURVE4,   # End of curve
    Path.CURVE4,
    Path.CURVE4,
    Path.CURVE4,
]

# Create path and patch
path = Path(verts, codes)
patch = patches.PathPatch(path, facecolor='none', lw=3)

# Plot
fig, ax = plt.subplots()
ax.add_patch(patch)
ax.set_xlim(-1, 5)
ax.set_ylim(-3, 3)
ax.set_aspect('equal')
plt.title("Cubic Bézier Curve")
plt.grid(True)
plt.show()
```
