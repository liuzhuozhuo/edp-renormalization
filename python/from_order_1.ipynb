Defining the canonical 1st order diagrams

```python
for i in range(len(can_points_1)):
    represent_diagram(can_points_1[i], can_paths_1[i], index = True)
```

From the canonical 1st order diagrams calculate the diagrams of 2nd order

```python
points_2, paths_2 = combine_diagrams_order([can_points_1], [can_paths_1], offset = 0)
for i in range(len(paths_2)):
    represent_diagram(points_2[i], trim_zeros_2D(paths_2[i]), index = True)
```

```python
points_3, paths_3 = combine_diagrams_order([can_points_1, points_2], [can_paths_1, paths_2], offset = 0.2)
n = 1
for i in range(len(paths_3)):
    in_out_paths_3 = in_out_paths(paths_3[i])
    if len(np.trim_zeros(in_out_paths_3[0])) == 1 and len(np.trim_zeros(in_out_paths_3[1])) == 2:
        represent_diagram(points_3[i], trim_zeros_2D(paths_3[i]), index = True, directory=f'diagrams/order3/{n}')
        n+=1
```

```python
points_4, paths_4 = combine_diagrams_order([can_points_1, points_2, points_3], [can_paths_1, paths_2, paths_3], offset = 0.2)

points_4, paths_4, count_4 = group_diagrams(points_4, paths_4)

n=1
for i in range(len(paths_4)):
    in_out_paths_4 = in_out_paths(paths_4[i])
    if len(np.trim_zeros(in_out_paths_4[0])) == 2 and len(np.trim_zeros(in_out_paths_4[1])) == 2:
        represent_diagram(points_4[i], trim_zeros_2D(paths_4[i]), index = True, number=count_4[i])
        n+=1
```
