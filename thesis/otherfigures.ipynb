```python
import numpy as np
import matplotlib.pyplot as plt

# Data from the table
orders = np.array([2, 3, 4, 5, 6])
unique_diagrams = np.array([16, 76, 612, 5871, 65000])
times = np.array([0.150, 0.198, 1.166, 10.935, 157.121])

# 1) Linear plot with twin y-axis
fig=plt.figure(figsize=(10,6)) #Size of the plot
ax1=fig.add_subplot(111)
line1, = ax1.plot(orders, unique_diagrams, marker='o', label='Types of diagrams', color='tab:orange')
ax1.set_xticks(orders)
ax1.set_xlabel('Order', fontsize=14)
ax1.set_ylabel('Types of diagrams', fontsize=14)
ax1.set_title('Types of Diagrams & Computational Time vs. Order', fontsize=14)

ax2 = ax1.twinx()
line2, = ax2.plot(orders, times, marker='s', label='Computational time (s)')
ax2.set_ylabel('Computational time (s)', fontsize=14)

# Legends
lines = [line1, line2]
labels = [line.get_label() for line in lines]
ax1.legend(lines, labels, loc='upper left', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data from the table
orders = np.array([2, 3, 4, 5, 6])
unique_diagrams = np.array([16, 76, 612, 5871, 65000])
times = np.array([0.150, 0.198, 1.166, 10.935, 157.121])

# Compute log10 values
log_unique = np.log(unique_diagrams)
log_times = np.log(times)

# Plot log₁₀ values with twin y-axes
fig=plt.figure(figsize=(10,6)) #Size of the plot
ax1=fig.add_subplot(111)
l1, = ax1.plot(orders, log_unique, marker='o', label='log(Types of diagrams)', color='tab:orange')
ax1.set_xticks(orders)
ax1.set_xlabel('Order', fontsize=14)
ax1.set_ylabel('log(Types of diagrams)', fontsize=14)
ax1.tick_params(axis='y')

ax2 = ax1.twinx()
l2, = ax2.plot(orders, log_times, marker='s', linestyle='-', label='log(Computational time)')
ax2.set_ylabel('log(Computational time)', fontsize=14)
ax2.tick_params(axis='y')

# Combined legend
lines = [l1, l2]
labels = [ln.get_label() for ln in lines]
ax1.legend(lines, labels, loc='upper left', fontsize=12)

ax1.set_title('Log Type of Diagrams & Computational Time vs. Order', fontsize=14)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```
