```python
import matplotlib.pyplot as plt

# Define coordinates (just use a list of (x, y) tuples)
coordinates = [(1, 1), (2, 3), (1, 5), (4, 3), (5, 5), (5, 1)]

# Auto-generate names: P, P, P, ...
names = [f"P{i}" for i in range(len(coordinates))]
positions = dict(zip(names, coordinates))

# Define undirected edges using the point names
edges = [('P0', 'P1'), ('P1', 'P2'), ('P1', 'P3'), ('P3', 'P4'), ('P3', 'P5')]

# Plot
plt.figure(figsize=(10, 6))
for name, (x, y) in positions.items():
    plt.plot(x, y, 'o', markersize=8, color='skyblue')
    plt.text(x + 0.1, y + 0.1, f"{name} ({x}, {y})", fontsize=14, color='tab:blue', ha='left', va='bottom')

for u, v in edges:
    x1, y1 = positions[u]
    x2, y2 = positions[v]
    plt.plot([x1, x2], [y1, y2], 'k-')
    mx, my = (x1 + x2) / 2, (y1 + y2) / 2
    plt.text(mx + 0.3, my + 0.1, f"[{u},{v}]", fontsize=14, color='black', ha='center', va='center')

plt.axis('off')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define coordinates
coordinates = [(1, 1), (2, 3), (1, 5), (4, 3), (5, 5), (5, 1)]

# Auto-generate names: P, P, ...
names = [f"{i}" for i in range(len(coordinates))]
positions = dict(zip(names, coordinates))

# Define undirected edges
edges = [('0', '1'), ('1', '2'), ('1', '3'), ('3', '4'), ('3', '5')]

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
for name, (x, y) in positions.items():
    plt.plot(x, y, 'o', markersize=8, color='tab:blue', zorder = 5)
    plt.text(x + 0.15, y + 0.1, f"{name} [{x}, {y}]", fontsize=14, color='tab:blue', ha='left', va='bottom', zorder = 5)

for u, v in edges:
    x1, y1 = positions[u]
    x2, y2 = positions[v]
    
    # Check for the special edge P1-P3 (or P3-P1)
    if (u == '1' and v == '3') or (u == '3' and v == '1'):
        plt.plot([x1, x2], [y1, y2], 'k--', linewidth=4)  # red dashed
    else:
        plt.plot([x1, x2], [y1, y2], 'k-', linewidth=4)
    
    mx, my = (x1 + x2) / 2, (y1 + y2) / 2
    plt.text(mx + 0.3, my + 0.15, f"[{u},{v}]", fontsize=14, color='black', ha='center', va='center')

plt.axis('off')
plt.tight_layout()
# Save the figure with a transparent background
output_path = "plots/example.png"
fig.savefig(output_path, transparent=True)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data from the table
orders = np.array([2, 3, 4, 5, 6])
unique_diagrams = np.array([16, 76, 612, 5871, 65000])
times = np.array([0.150, 0.198, 1.166, 10.935, 157.121])

# Compute log10 values
log_unique = np.log(unique_diagrams)
log_times = np.log(times)

# Plot log values with twin y-axes
fig=plt.figure(figsize=(10,6)) #Size of the plot
ax1=fig.add_subplot(111)
l1, = ax1.plot(orders, log_unique, marker='o', label='log(Types of diagrams)', color='tab:orange')
ax1.set_xticks(orders)
ax1.set_xlabel('Order', fontsize=14)
ax1.set_ylabel('log(Types of diagrams)', fontsize=14)
ax1.tick_params(axis='y')

ax2 = ax1.twinx()
l2, = ax2.plot(orders, log_times, marker='s', linestyle='-', label='log(Computational time)')
ax2.set_ylabel('log(Computational time)', fontsize=14)
ax2.tick_params(axis='y')

# Combined legend
lines = [l1, l2]
labels = [ln.get_label() for ln in lines]
ax1.legend(lines, labels, loc='upper left', fontsize=12)

ax1.set_title('Log Type of Diagrams & Computational Time vs. Order', fontsize=14)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Lightcone upper and lower surfaces
r = np.linspace(0, 2, 100)
theta = np.linspace(0, 2 * np.pi, 100)
R, Theta = np.meshgrid(r, theta)
Xc = R * np.cos(Theta)
Yc = R * np.sin(Theta)
Zc = R

# FF plane: x = t + z = const  z = x - t
x_plus_const = 1.5
T_plane = np.linspace(-2, 2, 100)
X_plane = np.linspace(-2, 2, 100)
T_grid, X_grid = np.meshgrid(T_plane, X_plane)
Z_grid = x_plus_const - T_grid

# Create figure
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# FIRST: plot bottom cone (behind the plane)
ax.plot_surface(Xc, Yc, -Zc, alpha=0.3, color='gray', linewidth=0)

# SECOND: plot the FF plane (in the middle)
ax.plot_surface(X_grid, T_grid, Z_grid, alpha=0.3, color='blue', linewidth=0)

# THIRD: plot top cone (in front of the plane)
ax.plot_surface(Xc, Yc, Zc, alpha=0.3, color='gray', linewidth=0)

# Axes settings
ax.set_xlabel('x')
ax.set_ylabel('t')
ax.set_zlabel('z')
ax.set_title("Lightcone and Front Form (FF) Plane: $x^+ = t + z = \\text{const}$")
ax.view_init(elev=25, azim=60)
plt.tight_layout()
plt.show()
```
